```python
from IPython.display import display, HTML
from google.colab.output import eval_js
```

```python
def set_html():
  html = HTML('''
    <!DOCTYPE html>
    <html lang="ja">
      <head>
        <meta charset="utf-8">
        <title>tic-tac-toe</title>
      </head>
      <body>
        <div id="board"></div>
        <span id="result"></span>
      </body>
      <style>
        button
        {
          width: 5em;
          height: 5em;
        }
      </style>
      <script>
        const state = {
          first: 0,
          second: 1,
          draw: 2,
          none: 3
        };
        let result = document.getElementById(`result`);
        let board = document.getElementById(`board`);
        let board_size;
        let buttons;
        let board_state;
        let play_first = true;
        let is_finished = false;

        function initialize(size)
        {
          board_size = size;
          for (let row = 0; row < board_size; row++) 
          {
            for (let column = 0; column < board_size; column++)
            {
              board.insertAdjacentHTML(`beforeend`, `<button onclick="play(${column + row * board_size});">`);
            }
            board.insertAdjacentHTML(`beforeend`, `<br>`);
          }
          buttons = document.getElementsByTagName(`button`);
          board_state = new Array(board_size * board_size);
          board_state.fill(0);
          update();
        }

        function play(index)
        {
          if (!is_correct(index))
          {
            return;
          }
          board_state[index] = play_first ? 1 : -1;
          play_first = !play_first;
          update();
        }

        function is_correct(index)
        {
          if (board_state[index] != 0)
          {
            return false;
          }
          if (is_finished)
          {
            return false;
          }
          return true;
        }

        function update()
        {
          for (let index = 0; index < board_size * board_size; index++)
          {
            if (board_state[index] == 1)
            {
              buttons[index].style.backgroundColor = `red`;
            }
            else if(board_state[index] == -1)
            {
              buttons[index].style.backgroundColor = `blue`;
            }
            else
            {
              buttons[index].style.backgroundColor = `white`;
            }
          }
          switch (judge())
          {
            case state.first:
              result.innerHTML = `赤の勝ち`;
              break;
            case state.second:
              result.innerHTML = `青の勝ち`;
              break;
            case state.draw:
              result.innerHTML = `引き分け`;              
              break;
          }
        }

        function judge()
        {
          let exists_empty = false;
          let column_score = new Array(board_size);          
          column_score.fill(0);
          let slash_score = 0;
          let back_slash_score = 0;
          for (let row = 0; row < board_size; row++) 
          {
            let row_score = 0;
            for (let column = 0; column < board_size; column++)
            {
              let current_state = board_state[column + row * board_size];
              if (current_state == 0)
              {
                exists_empty = true;
              }
              column_score[column] += current_state;
              row_score += current_state;
            }
            if (row_score == board_size)
            {
              return state.first;
            }
            if (row_score == -board_size)
            {
              return state.second;
            }
          }
          for (let column = 0; column < board_size; column++)
          {
            if (column_score[column] == board_size)
            {
              return state.first;
            }
            if (column_score[column] == -board_size)
            {
              return state.second;
            }
            back_slash_score += board_state[column + column * board_size];
            slash_score += board_state[column + (board_size - 1 - column) * board_size];
          }
          if (back_slash_score == board_size)
          {
            return state.first;
          }
          if (back_slash_score == -board_size)
          {
            return state.second;
          }
          if (slash_score == board_size)
          {
            return state.first;
          }
          if (slash_score == -board_size)
          {
            return state.second;
          }
          if (!exists_empty)
          {
            return state.draw;
          }
          return state.none;
        }

        function get_state(index)
        {
          return board_state[index];
        }
      </script>
    </html>
  ''')
  display(html)
```

```python
set_html()
eval_js('initialize({})'.format(5))
eval_js('play({})'.format(13))
value = eval_js('get_state({})'.format(13))
```
